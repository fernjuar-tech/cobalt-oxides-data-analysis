```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/cobalt_oxides.csv")

fig, axs = plt.subplots(1, 2, figsize=(10,5))

sns.scatterplot(data=df, x='ratioCoO', y='Eform1', hue='nH', ax=axs[0], palette="vlag")
sns.scatterplot(data=df, x='ratioCoO', y='Eform3', hue='nH', ax=axs[1], palette="coolwarm")

axs[0].set_title('O2/H2 reservoirs')
axs[1].set_title('H2/H2O reservoirs')

axs[0].set_xlabel('Co/O ratio')
axs[1].set_xlabel('Co/O ratio')

axs[0].set_ylabel('Formation Energy (eV)')
axs[1].set_ylabel('Formation Energy (eV)')

plt.savefig('figures/energies_withoutSOLV.png', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
